```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/Detailed-Sales-Report.xlsx',header=None)
df=df.iloc[2:,[2,4]]
df.columns=['Transaction','Item']
df
```

```python
grp=df.groupby('Transaction')['Item'].apply(list).reset_index()
grp
```

```python
df=grp['Item'].apply(pd.Series)
df
```

```python
#ECLAT_algorithm
from pyECLAT import ECLAT
eclat_instance=ECLAT(df,verbose=True)
```

```python
eclat_instance.uniq_
```

```python
get_eclat_indexes,get_eclat_supports=eclat_instance.fit(min_support=0.0001,min_combination=2,verbose=True,separator=' & ')
```
